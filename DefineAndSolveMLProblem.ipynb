```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Load data
airbnbDataSet_filename = os.path.join(os.getcwd(), "data", "airbnbListingsData.csv")
df = pd.read_csv(airbnbDataSet_filename)

# Print initial data info
print("===== Initial Data Info =====")
print(f"Shape: {df.shape}")
print("Columns:")
print(df.columns.tolist())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nPrice column stats:")
print(df['price'].describe())
print("-" * 40)

# Drop extreme outliers in 'price'
df = df[df['price'] <= 1000]

# Drop rows where price is missing
df = df.dropna(subset=['price'])

# Print price distribution after filtering
print("\nPrice stats after filtering extreme outliers:")
print(df['price'].describe())
print(f"Number of listings after filtering: {len(df)}")
print("-" * 40)

# Log-transform target
df['log_price'] = np.log1p(df['price'])

# Separate features and target
y = df['log_price']
X = df.drop(columns=['price', 'log_price'])

# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Split categorical columns by cardinality
low_card_cols = [col for col in categorical_cols if X[col].nunique() < 20]
high_card_cols = [col for col in categorical_cols if X[col].nunique() >= 20]

print("\nLow cardinality categorical columns:")
print(low_card_cols)
print("\nHigh cardinality categorical columns:")
print(high_card_cols)
print("-" * 40)

# Frequency encode high cardinality categoricals
for col in high_card_cols:
    freq = X[col].value_counts() / len(X)
    X[col + '_freq_enc'] = X[col].map(freq)

# Drop original high cardinality columns
X = X.drop(columns=high_card_cols)

# One-hot encode low cardinality categoricals
X_low_card = pd.get_dummies(X[low_card_cols], drop_first=True)

# Drop original low cardinality categorical columns
X_numeric = X.drop(columns=low_card_cols)

# Combine numeric + one-hot encoded categorical
X_encoded = pd.concat([X_numeric, X_low_card], axis=1)

# Print final feature info (no matrix print)
print("\nFinal feature matrix info:")
print(f"Shape: {X_encoded.shape}")
print(f"Number of features: {X_encoded.shape[1]}")
print("-" * 40)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Fill missing values with median of train set only (no leakage)
median_train = X_train.median()
X_train = X_train.fillna(median_train)
X_test = X_test.fillna(median_train)

# Initialize Random Forest with your original parameters
rf_model = RandomForestRegressor(
    n_estimators=500,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

# Train model
rf_model.fit(X_train, y_train)

# Predict on test set (log scale)
y_pred_log = rf_model.predict(X_test)

# Convert predictions back to original scale
y_test_exp = np.expm1(y_test)
y_pred_exp = np.expm1(y_pred_log)

# Create and print comparison table
comparison_df = pd.DataFrame({
    'Actual Price': y_test_exp.round(2),
    'Predicted Price': y_pred_exp.round(2)
})

print("\nSample of Actual vs Predicted Prices:")
print(comparison_df.head(15).to_string(index=False))

# Evaluate
r2 = r2_score(y_test_exp, y_pred_exp)
mae = mean_absolute_error(y_test_exp, y_pred_exp)
rmse = np.sqrt(mean_squared_error(y_test_exp, y_pred_exp))

print("\n===== Model Accuracy =====")
print(f"R^2 Score: {r2:.4f}")
print(f"Mean Absolute Error: ${mae:.2f}")
print(f"Root Mean Squared Error: ${rmse:.2f}")
print("-" * 40)

# Plot histogram of actual - predicted prices
errors = y_test_exp - y_pred_exp
plt.figure(figsize=(8, 6))
plt.hist(errors, bins=30, edgecolor='black', color='skyblue')
plt.axvline(0, color='red', linestyle='dashed', linewidth=2)
plt.xlabel("Actual Price - Predicted Price ($)")
plt.ylabel("Frequency")
plt.title("Histogram of Price Prediction Errors (Actual - Predicted)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot top 10 feature importances
importances = rf_model.feature_importances_
feature_names = X_train.columns
feat_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

top_n = 10
top_features = feat_importances.head(top_n)

print("\nTop 10 Features Affecting Price:")
print(top_features)

plt.figure(figsize=(10, 6))
top_features.plot(kind='barh', color='teal', edgecolor='black')
plt.xlabel("Importance Score")
plt.title(f"Top {top_n} Features Affecting Airbnb Price")
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()
```

We trained a Random Forest model to predict Airbnb prices using numeric and encoded categorical features, with the target log-transformed to reduce skewness. With 500 trees and full depth, the model achieved an RÂ² of about 0.60, explaining 60% of price variance. The mean absolute error was around $44,  and the RMSE was about $87, showing reasonable accuracy but some larger errors for expensive listings.

Predictions generally follow actual prices but can over or under estimate in the mid to high price range. The error distribution is roughly balanced around zero. Improvements could come from enhanced feature engineering, advanced models like gradient boosting, better outlier handling, and using cross validation for tuning. Overall, the model is a solid baseline but can be further optimized for better accuracy.
